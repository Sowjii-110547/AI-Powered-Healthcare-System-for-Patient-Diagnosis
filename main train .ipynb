```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier

import joblib



```

```python
# ---- BLOOD MODEL (multi-disease) ----
blood_df = pd.read_csv('Blood_samples_dataset_balanced_2(f).csv')

print(blood_df.head())
print(blood_df['Disease'].value_counts())

blood_features = [
    'Glucose',
    'Hemoglobin',
    'Platelets',
    'White Blood Cells',
    'Red Blood Cells',
    'BMI',
    'Systolic Blood Pressure',
    'Diastolic Blood Pressure',
    'Triglycerides',
    'HDL Cholesterol',
    'HbA1c',
    'Creatinine',
    'C-reactive Protein',
    'Heart Rate'
]

X_blood = blood_df[blood_features].values
y_blood = blood_df['Disease'].values

X_train_b, X_val_b, y_train_b, y_val_b = train_test_split(
    X_blood, y_blood, test_size=0.2, random_state=42, stratify=y_blood
)

blood_scaler = StandardScaler()
X_train_b_scaled = blood_scaler.fit_transform(X_train_b)
X_val_b_scaled = blood_scaler.transform(X_val_b)

blood_model = RandomForestClassifier(
    n_estimators=400,
    random_state=42,
    class_weight='balanced'
)
blood_model.fit(X_train_b_scaled, y_train_b)

print("BLOOD MODEL PERFORMANCE:")
print(classification_report(y_val_b, blood_model.predict(X_val_b_scaled)))
print("Blood model classes:", blood_model.classes_)

joblib.dump(blood_model, 'blood_model.pkl')
joblib.dump(blood_scaler, 'blood_scaler.pkl')
joblib.dump(blood_features, 'blood_feature_names.pkl')
```

```python
# ---- TEXT MODEL FOR "NONE" OPTION (NO CUSTOM CLASS) ----
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib

text_df = pd.read_csv('train.csv')[['Context']].dropna()
text_df.rename(columns={'Context': 'description'}, inplace=True)

X_text = text_df['description'].astype(str)

tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    stop_words='english'
)

# Fit only the vectorizer so we can transform future text
tfidf.fit(X_text)

# Save vectorizer and a constant label; no model object
joblib.dump(tfidf, 'text_tfidf.pkl')
joblib.dump('general', 'text_label.pkl')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import joblib

heart_df = pd.read_csv('Heart_Disease_Prediction.csv')
print(heart_df.columns)

# Label column
heart_label_col = 'Heart Disease'
print(heart_df[heart_label_col].value_counts())

# Choose numeric features (all exist in your columns list)
heart_features = [
    'Age',
    'BP',
    'Cholesterol',
    'Max HR',
    'ST depression'
]

X_heart = heart_df[heart_features].values
y_heart = heart_df[heart_label_col].values

Xh_train, Xh_val, yh_train, yh_val = train_test_split(
    X_heart, y_heart, test_size=0.2, random_state=42, stratify=y_heart
)

heart_scaler = StandardScaler()
Xh_train_s = heart_scaler.fit_transform(Xh_train)
Xh_val_s = heart_scaler.transform(Xh_val)

heart_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced'
)
heart_model.fit(Xh_train_s, yh_train)

print("HEART MODEL:")
print(classification_report(yh_val, heart_model.predict(Xh_val_s)))
print("Heart classes:", heart_model.classes_)

joblib.dump(heart_model, 'heart_model.pkl')
joblib.dump(heart_scaler, 'heart_scaler.pkl')
joblib.dump(heart_features, 'heart_feature_names.pkl')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import joblib

kid_df = pd.read_csv('kidney_disease_dataset.csv')
print(kid_df.columns)

# Label column
kidney_label_col = 'Target'
print(kid_df[kidney_label_col].value_counts())

# Choose some meaningful numeric features (all names copied from columns)
kidney_features = [
    'Age of the patient',
    'Blood pressure (mm/Hg)',
    'Random blood glucose level (mg/dl)',
    'Blood urea (mg/dl)',
    'Serum creatinine (mg/dl)',
    'Sodium level (mEq/L)',
    'Potassium level (mEq/L)',
    'Hemoglobin level (gms)',
    'Estimated Glomerular Filtration Rate (eGFR)'
]

X_k = kid_df[kidney_features].values
y_k = kid_df[kidney_label_col].values

Xk_train, Xk_val, yk_train, yk_val = train_test_split(
    X_k, y_k, test_size=0.2, random_state=42, stratify=y_k
)

kidney_scaler = StandardScaler()
Xk_train_s = kidney_scaler.fit_transform(Xk_train)
Xk_val_s = kidney_scaler.transform(Xk_val)

kidney_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced'
)
kidney_model.fit(Xk_train_s, yk_train)

print("KIDNEY MODEL:")
print(classification_report(yk_val, kidney_model.predict(Xk_val_s)))
print("Kidney classes:", kidney_model.classes_)

joblib.dump(kidney_model, 'kidney_model.pkl')
joblib.dump(kidney_scaler, 'kidney_scaler.pkl')
joblib.dump(kidney_features, 'kidney_feature_names.pkl')
```
